1. Email Validator (String Filtering + Loops)

```python
user_email = input("Enter your email: ")

has_at = False
for ch in user_email:
    if ch == "@":
        has_at = True

if has_at and (user_email.endswith(".com") or user_email.endswith(".org")):
    print("Valid Email")
else:
    print("Invalid Email")
```

2. Count Word Frequency in Customer Reviews

```python
sentence = input("Enter review:").lower()

words = sentence.split()
seen = []
pairs = []
res = "{"

for word in words:
    if word not in seen:
        count = words.count(word)
        pairs.append(f"'{word}': {count}")
        seen.append(word)

res += ", ".join(pairs) + "}"
print(res)
```

3. Password Strength Checker

```python
st = input("Enter password: ")

is_upper = False
is_lower = False
is_digit = False
is_special = False

for c in st:
    if c.isupper():
        is_upper = True
    elif c.islower():
        is_lower = True
    elif c.isdigit():
        is_digit = True
    elif not c.isalnum():
        is_special = True

if is_upper and is_lower and is_digit and is_special:
    print("Strong Password")
else:
    print("Not Strong Password")
```

4. Username Generator

```python
import random

st = input("Enter Name: ").lower().strip()

first_name = ""
last_name = ""
space_found = False

for ch in st:
    if ch == " " and not space_found:
        space_found = True
        continue
    if not space_found:
        first_name += ch
    else:
        last_name += ch

first3 = ""
last3 = ""

count = 0
for ch in first_name:
    if count < 3:
        first3 += ch
        count += 1

count = 0
for ch in range(len(last_name)-3, len(last_name)):
    if ch >= 0:
        last3 += last_name[ch]

final_st = first3 + last3 + str(random.randint(1,100))

print(final_st)
```

5. SMS Word Counter

```python
st = input("Enter Message: ")

word_count = len(st.split())

char_count = 0
status = "Within Limit"

for c in st:
    char_count += 1

if char_count > 160:
    status = "Limit Exceeded"

print(f"Words = {word_count}, Characters = {char_count}, Status = {status}")
```

6. Remove Duplicates from Contact List

```python
names = input("Enter names: ").split(",")

unique = []
for name in names:
    name = name.strip()
    clean = ""
    for ch in name:
        clean += ch
    if clean and clean.lower() not in [n.lower() for n in unique]:
        unique.append(clean)

print(", ".join(unique))
```

7. Detect Suspicious Words in Text (Moderation)

```python
user_input = input("Enter text: ").lower().split()

banned_words = ['spam', 'scam', 'fake']

for word in user_input:
    if word in banned_words:
        print("Warning: Suspicious content detected")
        break
else:
    print("No Suspicious content detected")
```

8. Invoice Code Formatter

```python
num = input("Enter the number: ").strip()

count = 0
for i in num:
    count += 1

zeros_needed = 5 - count
prefix = ""

for j in range(zeros_needed):
    prefix += "0"

print("INV" + prefix + num)
```

9. Count Hashtags and Mentions (Social Media Analytics)

```python
st = input("Enter Post: ")

count_hash = 0
count_at = 0

for c in st:
    if c == '#':
        count_hash += 1
    elif c == '@':
        count_at += 1

print(f"Hashtags = {count_hash}, Mentions = {count_at}")
```

10. Word Reversal for Data Encryption

```python
text = input("Enter Text: ").split()

rev_words = []


for word in text:  
    reversed_text = word[::-1]
    rev_words.append(reversed_text)


print(" ".join(rev_words))
```
