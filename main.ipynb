```python
user_email = input("Enter your email: ")

has_at = False
for ch in user_email:
    if ch == "@":
        has_at = True

if has_at and (user_email.endswith(".com") or user_email.endswith(".org")):
    print("Valid Email")
else:
    print("Invalid Email")
```

3. Password Strength Checker

```python
st = input("Enter password: ")

is_upper = False
is_lower = False
is_digit = False
is_special = False

for c in st:
    if c.isupper():
        is_upper = True
    elif c.islower():
        is_lower = True
    elif c.isdigit():
        is_digit = True
    elif not c.isalnum():
        is_special = True

if is_upper and is_lower and is_digit and is_special:
    print("Strong Password")
else:
    print("Not Strong Password")
```

4. Username Generator

```python
import random

st = input("Enter Name: ").lower()

new_st = ""
last_st = ""
final_st = ""

for i in range(3):
    new_st = new_st + st[i]

for j in range(-3,0):
    last_st = last_st + st[j]

final_st = new_st + last_st + str(random.randint(1,100))

print(final_st)
```

5. SMS Word Counter

```python
st = input("Enter Message: ")

word_count = len(st.split())

char_count = 0
status = "Within Limit"

for c in st:
    char_count += 1

if char_count > 160:
    status = "Limit Exceeded"

print(f"Words = {word_count}, Characters = {char_count}, Status = {status}")
```
